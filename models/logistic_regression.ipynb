```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cleaned/cleaned_reviews.csv")
df.head()
```

```python
X = df["clean_review"]      # cleaned text
y = df["sentiment"]         # positive or negative
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

print("Training size:", len(X_train))
print("Testing size:", len(X_test))
```

```python
tfidf = TfidfVectorizer(
    max_features=50000,
    ngram_range=(1,2),
    stop_words="english"
)

# Fit only on training data
X_train_vec = tfidf.fit_transform(X_train)

# Transform test data
X_test_vec = tfidf.transform(X_test)
```

```python
log_reg = LogisticRegression(
    max_iter=200,        # allow enough iterations
    n_jobs=-1            # use all CPU cores
)

log_reg.fit(X_train_vec, y_train)
```

```python
preds = log_reg.predict(X_test_vec)
```

```python
accuracy = accuracy_score(y_test, preds)
print("Logistic Regression Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, preds)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Logistic Regression Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print(classification_report(y_test, preds))
```

```python
sample = ["The movie was great but a little too long"]
sample_vec = tfidf.transform(sample)
print("Prediction:", log_reg.predict(sample_vec)[0])
```
